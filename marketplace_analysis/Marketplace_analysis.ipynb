#  Обзор бизнес-показателей маркетплейса

Маркетплейс находится в стадии активного роста и уже вышел на уровень устойчивой монетизации. Однако до настоящего момента продукт развивался преимущественно на основе бухгалтерских отчётов и интуитивных решений, без системной продуктовой аналитики.
С ростом бизнеса появилась необходимость глубже понимать поведение пользователей, источники роста, эффективность маркетинговых каналов и вклад отдельных категорий товаров в выручку. Выручка маркетплейса формируется в виде комиссии в размере 5% от стоимости каждой продажи.

В рамках данного аттестационного проекта проводится комплексный анализ продуктовых и бизнес-показателей маркетплейса: от привлечения и активации пользователей до монетизации и возврата инвестиций.

### Цель проекта

Необходимо провести комплексный анализ пользовательского поведения и ключевых бизнес-метрик маркетплейса, оценить эффективность монетизации и маркетинговых затрат, выявить инсайты и точки роста, а также подготовить основу для продуктовых гипотез и экспериментов.

### Задачи проекта

Для достижения цели в рамках проекта необходимо:
1. Первичный анализ и сбор данных.
2. Обзор ключевых метрик.
3. Оценка метрик монетизации и юнит-экономики.
4. Поиск инсайтов и точек роста. Сегментация и формулирование гипотез.
5. Подготовка эксперимента и подведение его итогов.

### 2.1 Эффективность монетизации

- Рассмотрим базовую бизнес-метрику **Revenue** (выручку), которая показывает растет ли бизнес, есть ли сезонность и нет ли резких провалов. В качестве выручки бизнеса используется комиссия в размере 5% от стоимости любой продажи.

````python
# Подготовим датасет
orders_df = pa_marketplace_orders_clean.copy()

orders_df['revenue'] = orders_df['total_price'] * 0.05
orders_df['order_date'] = orders_df['order_date'].dt.date
````

````python
# Revenue по дням
revenue_daily = (
    orders_df
    .groupby('order_date', as_index=False)
    .agg(revenue=('revenue', 'sum'))
)

display(revenue_daily.head())
````

````python
# Добавим скользящее среднее за 7 дней
revenue_daily['rolling_7d'] = (
    revenue_daily
    .sort_values('order_date')
    .rolling(7)
    .mean()
    .reset_index(level=0, drop=True)
)

# Построим график 
plt.figure(figsize=(12, 5))
plt.plot(revenue_daily['order_date'], revenue_daily['revenue'], label='Дневная динамика выручки')
plt.plot(revenue_daily['order_date'], revenue_daily['rolling_7d'], label='Скользящее среднее (7 дней)')
plt.title('Revenue (Выручка бизнеса)')
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.legend()
plt.grid()
plt.show()
````

В 2024 году выручка бизнеса быстро росла в первом полугодии, затем стабилизировалась и во второй половине года показала умеренное снижение. При высокой дневной волатильности сглаженная динамика не содержит резких провалов, что указывает на в целом стабильное состояние продукта.

---

- Чтобы разобраться, стали ли пользователи приносить меньше денег или их просто стало меньше, проанализируем **ARPU** и **ARPPU** в разрезе месяцев.

````python
# ARPU - средняя выручка на всех пользователей
orders_df['month'] = pd.to_datetime(orders_df['order_date']).dt.to_period('M').dt.start_time

arpu_arppu = (orders_df
        .groupby('month', as_index=False)
        .agg(revenue=('revenue', 'sum'),
            paying_users=('user_id', 'nunique'))
       )

# Добавим всех активных пользователей на этот месяц
arpu_arppu['total_users'] = (arpu_arppu['month']
                       .apply(lambda x: pa_marketplace_users[pa_marketplace_users['cohort_month'] <= x]['user_id'].nunique())
                      )

# Считаем ARPU и ARPPU
arpu_arppu['arpu'] = arpu_arppu['revenue'] / arpu_arppu['total_users']
arpu_arppu['arppu'] = arpu_arppu['revenue'] / arpu_arppu['paying_users']

display(arpu_arppu)
````

````python
# График ARPU
plt.figure(figsize=(12,5))
plt.plot(arpu_arppu['month'], arpu_arppu['arpu'])
plt.title('ARPU (выручка с каждого пользователя)')
plt.xlabel('Месяц')
plt.ylabel('ARPU')
plt.ylim(0, 300)
plt.grid()

# График ARPPU
plt.figure(figsize=(12,5))
plt.plot(arpu_arppu['month'], arpu_arppu['arppu'])
plt.title('ARPPU (выручка с каждого покупателя)')
plt.xlabel('Месяц')
plt.ylabel('ARPPU')
plt.ylim(0, 2000)
plt.grid()
plt.show()
````

ARPU на протяжении года резко снижается, а значит база пользователей растёт быстрее, чем их вклад в выручку. При этом ARPPU остаётся высоким и стабильным: платящие пользователи готовы тратить и видят ценность продукта. Маркетплейс умеет зарабатывать на платящих пользователях, но плохо превращает новых пользователей в покупателей. Основная точка роста - это улучшение онбординга и активации аудитории.

---

- Рассмотрим **AOV** - средний чек одного заказа. Это позволит понять внутреннюю механику продаж: стабильность ARPPU держится на дорогих товарах или на том, что пользователи стали чаще совершать мелкие покупки?

````python
# Рассчитаем AOV
aov = (
    orders_df
    .groupby('order_month', as_index=False)
    .agg(total_sales=('total_price', 'sum'),
         orders_count=('order_id', 'nunique'))
)

aov['aov'] = aov['total_sales'] / aov['orders_count']

display(aov)
````

````python
# Построим график
plt.figure(figsize=(12, 5))
plt.plot(aov['order_month'], aov['aov'])
plt.title('AOV (Средний чек одного заказа)')
plt.xlabel('Месяц')
plt.ylabel('Средний чек')
plt.ylim(0, 16000)
plt.grid()
plt.show()
````

Средний чек (AOV) на протяжении года демонстрирует стабильность, колеблясь в узком диапазоне около 14 000. Значит качество покупок не падает. Пользователи продолжают покупать товары той же ценовой категории. Проблема снижения выручки во втором полугодии связана не с тем, что старые клиенты стали экономить, а с тем, что их просто становится недостаточно для перекрытия растущей базы неактивных пользователей.

- Рассчитаем общую конверсию в заказ - **CR** по месяцам, чтобы увидеть реальную долю активных покупателей. Мы сопоставим количество реальных покупателей с общим числом активных пользователей (MAU), которых подробнее рассмотрим в следующем разделе.

````python
# Подготовим датасет
events_df = pa_marketplace_events.copy()
events_df['event_date'] = events_df['event_date'].dt.date
events_df['week'] = pd.to_datetime(events_df['event_date']).dt.to_period('W').dt.start_time
events_df['month'] = pd.to_datetime(events_df['event_date']).dt.to_period('M').dt.start_time

# Рассчитаем MAU
mau = events_df.groupby('month')['user_id'].nunique().reset_index(name='mau')

# Объединим данные о платящих пользователях и активных пользователях
cr = arpu_arppu[['month', 'paying_users']].copy()
cr = cr.merge(mau, on='month')

# Рассчитаем CR в заказ 
cr['cr'] = round((cr['paying_users'] / cr['mau']) * 100, 2)

display(cr)
````

````python
# Построим график 
plt.figure(figsize=(12, 5))
plt.plot(cr['month'], cr['cr'])
plt.title('CR (конверсия) из посещения в покупку по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Конверсия (%)')
plt.ylim(0, 50)
plt.grid()
plt.show()
````

Начиная с апреля, наблюдается устойчивый нисходящий тренд конверсии в покупку. Если весной конверсия достигала своего пика в 43.28%, то к декабрю она упала почти вдвое — до 24.07%.

Это говорит о том, что с каждым месяцем всё большая доля активных пользователей (MAU) не доходит до покупки. Привлекаемая аудитория либо менее целевая, либо сталкивается с барьерами, которые мешают совершить первый заказ.

**Вывод:**

- Метрики ARPPU и AOV остаются стабильно высокими. Это значит, что те пользователи, которые совершают покупки, лояльны к продукту, покупают на привычные суммы и приносят стабильный доход.

- Основной негативный тренд - резкое падение ARPU (с 226 до 30) и CR (с 43% до 24%). Это признак того, что рост базы пользователей происходит экстенсивно. Т.е. продукт не умеет удерживать интерес активных пользователей и конвертировать их в покупателей.

- Из-за падения конверсии рост аудитории перестал трансформироваться в рост выручки, что привело к её стабилизации и последующему плавному снижению во втором полугодии.


Текущая динамика указывает на наличие значительного потенциала в области активации и онбординга новых пользователей. Дальнейшее наращивание трафика без исправления воронки будет приводить к неэффективным тратам бюджета.

### 2.2 Активность и лояльность аудитории

- Рассмотрим **DAU/WAU/MAU, Stickiness** для изучения динамики активности пользователей и их вовлечённости в продукт.

````python
# Рассчитаем DAU/WAU
dau = events_df.groupby('event_date')['user_id'].nunique().reset_index(name='dau')
wau = events_df.groupby('week')['user_id'].nunique().reset_index(name='wau')
````

````python
# Построим графики 
plt.figure(figsize=(12, 5))
plt.plot(dau['event_date'], dau['dau'], label='Дневная активность')
plt.plot(wau['week'][:-1], wau['wau'][:-1], label='Недельная активность')
plt.plot(mau['month'], mau['mau'], label='Месячная активность')
plt.title('DAU/WAU/MAU (Активность пользователей)')
plt.xlabel('Дата')
plt.ylabel('Активные пользователи')
plt.legend()
plt.grid()
plt.show()
````

В начале года наблюдается активный рост всех метрик: DAU, WAU и MAU, что отражает успешное привлечение новых пользователей.
- Дневная активность (DAU) демонстрирует заметные колебания, характерные для маркетплейса с различной нагрузкой по дням недели. 

- Недельная (WAU) и месячная активность (MAU) имеют более сглаженный тренд: рост в первые месяцы года, пик в августе и умеренный спад во второй половине года. Для визуализации последняя неполная неделя WAU была исключена, что делает тренд более корректным.

Произведем расчёт метрики **Stickiness**.

````python
# Рассчитаем среднемесячные dau
dau['month'] = pd.to_datetime(dau['event_date']).dt.to_period('M').dt.start_time
dau_monthly = (dau
               .groupby('month', as_index=False)
               .agg(avg_dau=('dau', 'mean'))
               )

# Расчёт Stickness
stick = pd.merge(mau, dau_monthly, how='left', on='month')
stick['stick'] = round((stick['avg_dau'] / stick['mau']) * 100, 2)

display(stick)
````

````python
# Построим графики
plt.figure(figsize=(12, 5))
plt.plot(stick['month'], stick['stick'])
plt.title('Stickness (Cпособность продукта удерживать пользователей)')
plt.xlabel('Месяц')
plt.ylabel('Stickiness, %')
plt.grid()
plt.show()
````

Показатель Stickiness демонстрирует умеренную, но устойчивую тенденцию к снижению во втором полугодии. 
Это означает, что пользователи стали реже заходить в приложение. 

Снижение вовлеченности во втором полугодии по времени совпадает с падением конверсии в заказ и ARPU. Это подтверждает, что новые пользователи не только меньше покупают, но и в целом меньше интересуются продуктом после первого знакомства.

Нисходящий тренд Stickiness указывает на то, что приток новой аудитории во второй половине года менее лоялен или менее заинтересован в функционале платформы.

Несмотря на то, что MAU удерживается на высоком уровне, качественные показатели активности (Stickiness) падают вслед за показателями монетизации. Это еще раз подчеркивает необходимость работы над активацией, чтобы превратить заходящую аудиторию в вовлеченных пользователей.

### 2.3 Юнит-экономика и маркетинг

- Проанализируем метрики **CAC, ROI, LTV** и **Payback Period**, чтобы оценить эффективность маркетинга: стоимость привлечения клиента, его способность окупить маркетинговые затраты за счёт комиссии (5%) и общую отдачу от вложений в продвижение.

````python
# Рассчитаем маркетинговые затраты по месяцам
marketing_monthly = (
    pa_marketplace_campaign_costs
    .groupby('spend_month', as_index=False)
    .agg(total_budget=('budget', 'sum'))
)

# Объединяем с данными о монетизации
unit_economic = arpu_arppu.merge(marketing_monthly, left_on='month', right_on='spend_month', how='left')

# Расчет CAC
unit_economic['cac'] = unit_economic['total_budget'] / unit_economic['paying_users']

# Расчет ROI
unit_economic['roi'] = (unit_economic['revenue'] - unit_economic['total_budget']) / unit_economic['total_budget'] * 100

# Расчет LTV упрощенный
unit_economic['ltv'] = unit_economic['revenue'] / unit_economic['paying_users']

display(unit_economic[['month', 'total_budget', 'cac', 'ltv', 'roi']].round(2))
````

````python
# Графикb CAC и LTV
plt.figure(figsize=(12,5))
plt.plot(unit_economic['month'], unit_economic['cac'], label='CAC')
plt.plot(unit_economic['month'], unit_economic['ltv'], label='LTV')
plt.title('CAC (Стоимость привлечения одного клиента) \nи LTV (Прибыль от клиента за всё время взаимодействия)')
plt.xlabel('Месяц')
plt.ylabel('Значения')
plt.ylim(0, 2700)
plt.legend()
plt.grid()


# График ROI
plt.figure(figsize=(12,5))
plt.plot(unit_economic['month'], unit_economic['roi'])
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.title('ROI (Коэффициент возврата инвестиций)')
plt.xlabel('Месяц')
plt.ylabel('ROI %')
plt.grid()
plt.show()
````

**Вывод:**

- С начала года стоимость привлечения платящего пользователя (CAC) выросла почти в 4 раза (с 866 в апреле до 2314 в декабре). В то же время доход с одного клиента (LTV) остался стабильным (на уровне ~1600–1700). В октябре кривая затрат пересекла кривую доходности, и с этого момента каждый новый привлеченный клиент приносит компании прямой убыток.


- Показатель окупаемости инвестиций продемонстрировал падение (с пиковых 100.56% в мае до -28.3% в декабре). Это означает, что в конце года на каждый вложенный в рекламу рубль компания теряет 28 копеек, так как 5% комиссии не покрывают маркетинговые расходы.

---

- Рассчитаем **Payback Period** (срок окупаемости) - показать, сколько средних заказов нужно совершить пользователю, чтобы покрыть свой CAC.

````python
# Добавим расчет AOV в таблицу unit_economic
unit_economic = unit_economic.merge(aov[['order_month', 'aov']], left_on='month', right_on='order_month', how='left')

# Рассчитаем, какую прибыль приносит один средний заказ (5% от чека)
unit_economic['profit_per_order'] = unit_economic['aov'] * 0.05

# Рассчитаем Payback Period 
unit_economic['payback_orders'] = unit_economic['cac'] / unit_economic['profit_per_order']

display(unit_economic[['month', 'cac', 'profit_per_order', 'payback_orders']].round(2))
````

- ARPPU в разрезе сегментов

- Долю выручки, которую генерирует каждый сегмент.

````python
# Добавим флаг по типу пользователя для каждого месяца
events_users_type = events_df.merge(pa_marketplace_users[['user_id', 'cohort_month']], on='user_id', how='inner')
orders_users_type = orders_df.merge(pa_marketplace_users[['user_id', 'cohort_month']], on='user_id', how='inner')

# Флаг пользователя: 1-новый пользователь, 0-возвращающийся
events_users_type['new_user'] = (events_users_type['event_month'] == events_users_type['cohort_month']).astype(int)
orders_users_type['new_user'] = (orders_users_type['order_month'] == orders_users_type['cohort_month']).astype(int)

# Рассчитаем MAU по типам пользователей
metrics_users_type = (events_users_type
           .groupby(['new_user', 'month'])['user_id']
           .nunique()
           .reset_index(name='mau'))

# Рассчитаем Paying Users и Revenue
paying_users_type = (orders_users_type
               .groupby(['new_user', 'month'], as_index=False)
               .agg(paying_users = ('user_id', 'nunique'),
                    total_sales = ('total_price', 'sum')))

paying_users_type['revenue'] = paying_users_type['total_sales'] * 0.05

# Объединяем с metrics 
metrics_users_type = (metrics_users_type
                      .merge(paying_users_type[['new_user', 'month', 'paying_users', 'revenue']], 
                             on=['new_user', 'month'], how='left'))

# Считаем CR
metrics_users_type['cr'] = round(metrics_users_type['paying_users'] / metrics_users_type['mau'] * 100, 2)

# Считаем ARPPU
metrics_users_type['arppu'] = metrics_users_type['revenue'] / metrics_users_type['paying_users']

# Доля выручки сегмента в месяце
total_rev_month = (metrics_users_type
                   .groupby('month', as_index=False)
                   .agg(total_rev_month = ('revenue', 'sum')))
                   
# Присоединяем к metrics_users_type
metrics_users_type = metrics_users_type.merge(total_rev_month, on='month', how='left')

# Доля выручки сегмента
metrics_users_type['revenue_share'] = (
    metrics_users_type['revenue'] / metrics_users_type['total_rev_month']
) * 100

display(metrics_users_type.head().round(2))
````

- Поскольку наши данные ограничены 2024 годом, в январе мы видим только новых пользователей. Для более корректного анализа динамики формирования лояльного ядра и сравнения сегментов, на графиках ниже мы рассмотрим период, начиная с февраля.

### 4.3 Сегментация по устройствам 

Чтобы выявить возможные технические барьеры на пути пользователя к покупке проверим, насколько удобно совершать заказы с разных платформ. Наша задача - понять, как платформа влияет на желание и возможность совершить покупку.

Проанализируем:

- нет ли резкого падения конверсии на конкретной ОС, что указывало бы на ошибки в коде.

- с чего пользователи покупают эффективнее - с мобильных устройств, компьютеров или планшетов.

- как отличается ARPPU у пользователей разных платформ (например, iOS и macOS против Android и Windows).

````python
# Определяем основной девайс и ОС для каждого пользователя
users_device = (events_df
                 .groupby(['user_id', 'device', 'os'], as_index=False)
                 .agg(event_count = ('event_id', 'count'))
                 .sort_values(['user_id', 'event_count'], ascending=[True, False])
                 .drop_duplicates('user_id'))

# Добавим информацию об устройстве пользователя к событиям и заказам
events_device = ((events_df.drop(columns=['device', 'os']))
                 .merge(users_device[['user_id', 'device', 'os']], on='user_id', how='left'))
orders_device = (orders_df
                 .merge(users_device[['user_id', 'device', 'os']], on='user_id', how='left'))

# Рассчитаем MAU по типам устройств
metrics_device = (events_device
           .groupby(['device', 'os', 'month'])['user_id']
           .nunique()
           .reset_index(name='mau'))

# Рассчитаем Paying Users и Revenue
paying_device = (orders_device
               .groupby(['device', 'os', 'month'], as_index=False)
               .agg(paying_users = ('user_id', 'nunique'),
                    total_sales = ('total_price', 'sum')))

paying_device['revenue'] = paying_device['total_sales'] * 0.05

# Объединяем с metrics 
metrics_device = (metrics_device
                      .merge(paying_device[['device', 'os', 'month', 'paying_users', 'revenue']], 
                             on=['device', 'os', 'month'], how='left'))

# Считаем CR
metrics_device['cr'] = round(metrics_device['paying_users'] / metrics_device['mau'] * 100, 2)

# Считаем ARPPU
metrics_device['arppu'] = metrics_device['revenue'] / metrics_device['paying_users']

display(metrics_device.head().round(2))
````

````python
# Агрегируем MAU по устройствам
mau_device = metrics_device.groupby(['month', 'device'])['mau'].sum().reset_index()
# Агрегируем MAU по ОС
mau_os = metrics_device.groupby(['month', 'os'])['mau'].sum().reset_index()

# Построим график MAU по устройствам
plt.figure(figsize=(12,6))
sns.lineplot(data=mau_device, x='month', y='mau', hue='device')
plt.title('MAU по устройствам')
plt.ylabel('MAU')
plt.xlabel('Месяц')
plt.grid()

# Построим график MAU по ОС
plt.figure(figsize=(12,6))
sns.lineplot(data=mau_os, x='month', y='mau', hue='os')
plt.title('MAU по ОС')
plt.ylabel('MAU')
plt.xlabel('Месяц')
plt.grid()
plt.show()
````

Трафик равномерно распределяется между устройствами и ОС, без явных лидеров. Все сегменты росли до августа 2024 года и затем вышли на плато с незначительным снижением к концу года, что указывает на общую тенденцию, а не проблемы конкретной платформы.

---

- Рассмотрим динамику CR по устройствам, чтобы понять, где пользователям удобнее покупать.

````python
# Агрегируем CR по устройствам
cr_device = metrics_device.groupby(['month', 'device'])[['paying_users', 'mau']].sum().reset_index()
cr_device['cr'] = (cr_device['paying_users'] / cr_device['mau']) * 100

# Построим график MAU по устройствам
plt.figure(figsize=(12,6))
sns.lineplot(data=cr_device, x='month', y='cr', hue='device')
plt.title('CR по устройствам')
plt.ylabel('CR, %')
plt.xlabel('Месяц')
plt.ylim(0)
plt.grid()
plt.show()
````

В первой половине года Mobile демонстрировал лучший CR ( в мае), однако к концу года эффективность всех платформ сравнялась на низком уровне около 20%.

Параллельное движение графиков `Desktop`, `Mobile` и `Tablet` указывает на то, что интерфейс везде работает стабильно, и критических преград для покупки нет.

- Рассмотрим динамику ARPPU по ОС. Традиционно ожидается, что `iOS` и `macOS` приносят больше дохода, чем `Windows` или `Android`.

В ходе анализа мы изучим:

- Активность пользователей (MAU): Поймем, в каких регионах сосредоточен основной трафик и где произошел наибольший приток аудитории во втором полугодии.


- Структуру выручки: Определим ключевые регионы по общему доходу.


- Эффективность (CR и ARPPU): С помощью тепловых карт выявим аномальные зоны и проверим синхронность изменения метрик по всей географии присутствия.

````python
# Объединяем события и заказы с данными о регионах 
events_geo = events_df.merge(pa_marketplace_users[['user_id', 'region']], on='user_id', how='left')
orders_geo = orders_df.merge(pa_marketplace_users[['user_id', 'region']], on='user_id', how='left')

# Рассчитываем ключевые метрики по сегментам и месяцам
metrics_geo = (events_geo
               .groupby(['region', 'month'], as_index=False)
               .agg(mau=('user_id', 'nunique')))

paying_geo = (orders_geo
                   .groupby(['region', 'month'], as_index=False)
                   .agg(paying_users=('user_id', 'nunique'),
                        total_sales=('total_price', 'sum')))

# Объединяем данные
metrics_geo = metrics_geo.merge(paying_geo, on=['region', 'month'], how='left')

metrics_geo['revenue'] = metrics_geo['total_sales'] * 0.05
metrics_geo['cr'] = metrics_geo['paying_users'] / metrics_geo['mau'] * 100
metrics_geo['arppu'] = metrics_geo['revenue'] / metrics_geo['paying_users']

display(metrics_geo.head().round(2))
````

````python
# Топ-5 регионов по MAU
top_regions_mau = (metrics_geo
               .groupby('region', as_index=False)['mau'].sum()               
               .sort_values('mau', ascending=False)
               .head(5))

display(top_regions_mau)

# Построим график 
plt.figure(figsize=(12, 5))
bars_region = plt.bar(top_regions_mau['region'], top_regions_mau['mau'])

# Перебираем в цикле значения, которые мы только что отобразили
for bar in bars_region:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', fontsize=12)

plt.title('Топ-5 регионов по MAU')
plt.xlabel('Регион')
plt.ylabel('Количество уникальных пользователей')
plt.show()
````

````python
# Топ-5 регионов по выручке
top_regions = (metrics_geo
               .groupby('region', as_index=False)['revenue'].sum()               
               .sort_values('revenue', ascending=False)
               .head(5))

display(top_regions)

# Построим график 
plt.figure(figsize=(12, 5))
bars_region = plt.bar(top_regions['region'], top_regions['revenue'])

# Перебираем в цикле значения, которые мы только что отобразили
for bar in bars_region:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', fontsize=12)

plt.title('Топ-5 регионов по выручке')
plt.xlabel('Регион')
plt.ylabel('Выручка')
plt.show()
````

````python
# Преобразуем дату в формат '2024-01-01'
metrics_geo['month'] = metrics_geo['month'].dt.strftime('%Y-%m-%d')

# Подготовка данных для Heatmap 
cr_pivot = metrics_geo.pivot(index="region", columns="month", values="cr")
arppu_pivot = metrics_geo.pivot(index="region", columns="month", values="arppu")

# Тепловая карта для CR
plt.figure(figsize=(12, 5))
sns.heatmap(cr_pivot, annot=True, fmt=".1f", cbar_kws={'label': 'Конверсия, %'})
plt.title('Тепловая карта CR по регионам и месяцам')
plt.xlabel('Месяц')
plt.ylabel('Регион')

# Тепловая карта для ARPPU
plt.figure(figsize=(12, 5))
sns.heatmap(arppu_pivot, annot=True, fmt=".0f", cbar_kws={'label': 'ARPPU'})
plt.title('Тепловая карта ARPPU по регионам и месяцам')
plt.xlabel('Месяц')
plt.ylabel('Регион')


plt.show()
````

### 4.5 Анализ воронки

Для поиска слабых мест мы выстроим события в цепочку, отражающую реальный путь пользователя к покупке. 

Этапы воронки:
1. Просмотр платформы (`page_view`).
2. Интерес к товару (`product_view`).
3. Намерение купить (`add_to_cart`).
4. Начало оформления (`checkout_start`).
5. Завершение оформления заказа (`checkout_complete`).
5. Оплата (данные по оплате из таблицы `orders_df`).

````python
# Задаем порядок шагов в воронке
funnel_steps = ['page_view', 'product_view', 'add_to_cart', 'checkout_start', 'checkout_complete']

# Фильтруем датасет оставляя только нужные шаги воронки
funnel_df_steps = events_df[events_df['event_type'].isin(funnel_steps)]

# Считаем число уникальных пользователей на каждом шаге
funnel_counts = (
    funnel_df_steps.groupby('event_type')['user_id']
    .nunique()
    .reindex(funnel_steps)
)

# Считаем количество уникальных покупателей из таблицы заказов
paying_users_count = orders_df['user_id'].nunique()

# Собираем таблицу воронки
funnel_df = pd.DataFrame({
    'step':funnel_counts.index.tolist() + ['payment_confirmed'],
    'users':funnel_counts.values.tolist() + [paying_users_count]
})

# Добавляем конверсии и округляем до 1 знака после запятой
funnel_df['conversion_from_first_%'] = funnel_df['users'] / funnel_df['users'][0] * 100

funnel_df['conversion_from_previous_%'] = funnel_df['users'] / funnel_df['users'].shift(1) * 100

display(funnel_df.round(2))
````

````python
steps = funnel_df['step'][1:]
conversions_classic = funnel_df['conversion_from_first_%'][1:]
conversions_by_step = funnel_df['conversion_from_previous_%'][1:]

# Визуализируем данные классической воронки
plt.figure(figsize=(12, 5))

# Строим barplot
plt.bar(steps,
    conversions_classic)

# Перебираем в цикле значения конверсии, которые мы только что отобразили
for index, value in enumerate(conversions_classic):
    # Для каждого значения добавляем текст на график
    
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', fontsize=12, color='black')

plt.title('Классическая воронка: доля от первого шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')

# Визуализируем данные воронки step-by-step
plt.figure(figsize=(12, 5))

# Строим barplot
plt.bar(steps, 
        conversions_by_step)

# Перебираем в цикле значения конверсии, которые мы только что отобразили
for index, value in enumerate(conversions_by_step):
    # Для каждого значения добавляем текст на график
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', fontsize=12, color='black')

plt.title('Step-by-step воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')

plt.show()
````

Визуализация классической и пошаговой воронки позволяет сделать следующие выводы:

- Переход от добавления в корзину к началу оформления показывает заметное снижение (82.2%). Суммарно от первого шага до оплаты доходит лишь 40.4% пользователей.


- Самый резкий обрыв воронки происходит на последнем шаге. Конверсия из завершения оформления (`checkout_complete`) в реальную оплату (`payment_confirmed`) составляет всего 64.2%. Это означает, что каждый третий пользователь, полностью заполнивший данные для заказа, в итоге не совершает покупку. Это может быть связано с техническими ошибками, отсутствием удобных способов оплаты или скрытыми комиссиями, которые пользователь видит в самый последний момент.

---

Общая воронка выявила критическую зону - этап перехода от завершения оформления к подтверждению оплаты. 
Ранее мы заметили аномальную активность в канале `TikTok`. Чтобы локализовать проблему, мы проведем сравнительный анализ Step-by-step воронки канала `TikTok` с остальными каналами.

````python
# Помечаем пользователей: TikTok или Other
events_channels_df['group'] = events_channels_df['acq_channel'].apply(lambda x: 'TikTok' if x == 'TikTok' else 'Other')
orders_channels_df['group'] = orders_channels_df['acq_channel'].apply(lambda x: 'TikTok' if x == 'TikTok' else 'Other')

# Считаем шаги из логов по группам
funnel_counts = (events_channels_df[events_channels_df['event_type'].isin(funnel_steps)]
    .groupby(['group', 'event_type'])['user_id']
    .nunique()
)

# Считаем оплаты по группам
paying_users = orders_channels_df.groupby('group')['user_id'].nunique()

# Создаем отдельные Series для каждой группы
tiktok_df = funnel_counts['TikTok'].reindex(funnel_steps).tolist() + [paying_users['TikTok']]
other_df = funnel_counts['Other'].reindex(funnel_steps).tolist() + [paying_users['Other']]

# Собираем итоговый результат для вывода
final_table = pd.DataFrame({
    'tiktok': tiktok_df,
    'other': other_df
}, index=funnel_steps + ['payment_confirmed'])

# Считаем Step-by-step конверсию (от предыдущего шага)
group_table = pd.DataFrame()
group_table['tiktok_step_%'] = (final_table['tiktok'] / final_table['tiktok'].shift(1) * 100)
group_table['other_step_%'] = (final_table['other'] / final_table['other'].shift(1) * 100)

display(group_table.round(2).iloc[1:])
````

````python
# Берём только нужные строки 
plot_df = group_table.iloc[1:].round(2)

x = np.arange(len(plot_df.index))  
width = 0.35  

plt.figure(figsize=(10, 5))

plt.bar(x - width/2, plot_df['other_step_%'], width, label='Other')
plt.bar(x + width/2, plot_df['tiktok_step_%'], width, label='TikTok')

# Добавляем подписи значений
for i, val in enumerate(plot_df['other_step_%']):
    plt.text(i - width/2, val + 1, f'{val:.0f}',
             ha='center', fontsize=9)

for i, val in enumerate(plot_df['tiktok_step_%']):
    plt.text(i + width/2, val + 1, f'{val:.0f}',
             ha='center', fontsize=9)

plt.xticks(x, plot_df.index, rotation=45)
plt.ylabel('Conversion %')
plt.title('Step-by-step воронка канала TikTok и остальных')
plt.legend()
plt.show()
````

Сравнительный анализ воронки подтвердил, что канал TikTok является определяющим негативный тренд всего маркетплейса:

- Конверсия из оформления в оплату у `TikTok` составляет всего 27%, что почти в 3 раза ниже, чем у остальных каналов (73%). Столь резкий обвал на финальном шаге указывает на импульсивный характер аудитории, которая легко доходит до чекаута, но массово отказывается от реальных трат.


- Общее падение CR маркетплейса вызвано не поломкой продукта, а изменением структуры трафика в пользу некачественного сегмента из `TikTok`.

**Вывод по анализу воронки:**

Детальное изучение пути пользователя позволило локализовать причину падения общей конверсии маркетплейса:

### 4.6 Категории товаров

Чтобы понять, какие товары привлекают пользователей и где мы теряем прибыль, рассмотрим распределение заказов и выручки по категориям. 

- Чтобы понять структуру продаж, нам нужно посмотреть на две метрики: объем выручки и конверсию из просмотра в покупку для всего маркетплейса.

````python
# Добавляем категорию товара в таблицу событий
product_category_dict = orders_channels_df[['product_name', 'category_name']].drop_duplicates()

events_with_cat = events_channels_df.merge(product_category_dict, on='product_name', how='left')

# Считаем общие просмотры по категориям 
total_cat_views = (
    events_with_cat[events_with_cat['event_type'] == 'product_view']
    .groupby('category_name')['user_id']
    .nunique()
    .reset_index(name='total_viewers')
)

# Считаем общие покупки и выручку
total_cat_sales = (
    orders_channels_df
    .groupby('category_name')
    .agg({'user_id': 'nunique', 'total_price': 'sum'})
    .reset_index()
    .rename(columns={'user_id': 'total_buyers', 'total_price': 'revenue'})
)

# Объединяем и считаем общую конверсию
total_category_df = total_cat_views.merge(total_cat_sales, on='category_name', how='left').fillna(0)
total_category_df['conversion_%'] = (total_category_df['total_buyers'] / total_category_df['total_viewers'] * 100).round(2)

# Сортируем по выручке, чтобы увидеть самые важные категории
total_category_rev = total_category_df.sort_values(by='revenue', ascending=False)
total_category_conv = total_category_df.sort_values(by='conversion_%', ascending=False)

display(total_category_rev)
````

````python
# Построим график выручки
plt.figure(figsize=(10, 5))
top_rev = total_category_rev.head(10)
sns.barplot(data=top_rev, x='category_name', y='revenue', color='steelblue')

# Добавляем подписи
for i, val in enumerate(top_rev['revenue']):
    plt.text(i, val, f'{val/1e6:.1f}M', ha='center', va='bottom', fontsize=10)

plt.title('Топ-10 категорий по выручке')
plt.xticks(rotation=45)
plt.ylabel('Выручка')
plt.show()

# Построим график конверсии
plt.figure(figsize=(10, 5))
top_conv = total_category_conv.head(10)
sns.barplot(data=top_conv, x='category_name', y='conversion_%', color='skyblue')

# Добавляем подписи
for i, val in enumerate(top_conv['conversion_%']):
    plt.text(i, val + 0.5, f'{val}%', ha='center', va='bottom', fontsize=10)

plt.title('Топ-10 категорий по конверсии')
plt.xticks(rotation=45)
plt.ylim(0, top_conv['conversion_%'].max() + 5)
plt.ylabel('Конверсия (%)')
plt.show()
````

На основе полученных данных можно выделить следующие ключевые закономерности:

- Основной объем доходов маркетплейса формируют категории `Мебель для дома` (почти 70 млн) и `Бытовая техника` (около 43 млн). При этом их конверсия находится на среднем уровне 13.23% и 14.21% соответственно.


- Категория `Продукты питания` демонстрирует рекордную конверсию (67.47%). Несмотря на то, что она не входит в топ по выручке, это самый эффективный сегмент с точки зрения завершения сделки.


- Категории `Детская одежда` (23.69%) и `Зоотовары` (22.53%) показывают конверсию выше среднего при умеренной выручке. Это указывает на высокую лояльность или понятную ценность товара для покупателя в этих нишах.


- Категории `Украшения и часы` и `Товары для туризма` генерируют значительную выручку (33 млн и 22 млн), но имеют относительно невысокую конверсию - около 12.7%. Незначительное увеличение CR в этих сегментах может дать существенный прирост в выручке.

Общая статистика показывает, что у маркетплейса есть лидеры:

- **Бизнес-контекст:** Команда продукта запускает эксперимент, направленный на повышение конверсии среди пользователей, привлечённых через канал `TikTok`. Для стимулирования первой покупки пользователям будет предоставлен бонус на первую покупку.
- **Цель:** Проверить, влияет ли предоставление бонуса на первую покупку на конверсию и выручку среди пользователей `TikTok`.
- **Гипотеза:** Предоставление бонуса на первую покупку новым пользователям, привлеченным через `TikTok`, приведет к увеличению конверсии в первую покупку
---
#### 2. Формат эксперимента
- **Тип эксперимента:** Сплит-роллинг (rolling split)
- **Канал:** Только пользователи, пришедшие с `TikTok`
- **Группы:**
    - **Контрольная (A):** пользователи `TikTok` без бонуса
    - **Тестовая (B):** пользователи `TikTok` с бонусом на первую покупку
- **Распределение:** Равномерное распределение новых пользователей `TikTok` между группами по дате регистрации
- **Период проведения:** 3 недели (фиксировано)
---
#### 3. Ключевые метрики

##### Основная метрика
- **Конверсия в первую покупку** — доля пользователей, зарегистрировавшихся с TikTok и совершивших первую покупку
- *Цель:* увеличить конверсию с помощью бонуса
- *Критерий успеха:* статистически значимый рост конверсии в тестовой группе

##### Вспомогательные метрики
- **ARPU** (средняя выручка на пользователя TikTok)
- **ARPPU** (средняя выручка на покупателя TikTok)
- **AOV** (средний чек TikTok)

##### Барьерная метрика
- **Общая выручка от пользователей TikTok**
---
#### 4. Размеры групп и запуск
- **Целевая аудитория:** только пользователи TikTok
- **Размеры групп:** определяются по фактическому трафику TikTok за период эксперимента
- **Комментарий:** запуск без предварительного расчёта мощности и MDE из-за срочности задачи

### 5.3 Анализ метрик А/В-эксперимента 

Проанализируем такие метрики:

**Основная метрика**

- Конверсия в первую покупку — доля пользователей, зарегистрировавшихся с TikTok и совершивших первую покупку.

Цель: увеличить конверсию с помощью бонуса.

Критерий успеха: статистически значимый рост конверсии в тестовой группе.


**Вспомогательные метрики**

- ARPU (средняя выручка на пользователя TikTok).

- ARPPU (средняя выручка на покупателя TikTok).

- AOV (средний чек TikTok).


**Барьерная метрика**

- Общая выручка от пользователей TikTok.

Для каждой метрики выведем: 
- значения в контрольной и тестовой группах, 
- абсолютные и относительные изменения значений в тестовой группе по сравнению с контрольной.

После этого проверим статистическую значимость различий метрик между группами эксперимента. 

````python
# Считаем уникальных пользователей в каждой группе
users_per_group = df_users_AB.groupby('group')['user_id'].nunique()

# Считаем покупателей (совершивших хотя бы один заказ)
buyers_per_group = df_orders_AB.groupby('group')['user_id'].nunique()

# Конверсия
conversion = (buyers_per_group / users_per_group * 100).round(2)
conversion_change = ((conversion['treatment'] - conversion['control']) / conversion['control'] * 100).round(2)

print("Конверсия в первую покупку (%):")
display(conversion)
print(f"Относительное изменение в тестовой группе: {conversion_change}%")
````

Конверсия в первую покупку в тестовой группе составила 7.51%, тогда как в контрольной — лишь 1.06%.
Относительный прирост конверсии достиг +608%, что указывает на сильное положительное влияние бонуса на поведение пользователей, привлечённых через `TikTok`.

Однако столь большой эффект требует дополнительной проверки статистической значимости различий между группами, чтобы исключить влияние случайных колебаний выборки.

- Для проверки значимости различий в конверсии применим z-тест для долей, который позволяет корректно сравнить бинарную метрику (совершил покупку / не совершил) между двумя независимыми группами.

**Нулевая гипотеза (H0):**
Конверсия в первую покупку одинаковая в тестовой и контрольной группах.

**Альтернативная гипотеза (H1):**
Конверсия в первую покупку в тестовой и контрольной группах различается.

````python
# Подготовка датафрейма с флагом первой покупки
first_purchase_df = df_users_AB[['user_id', 'group']].copy()

# Флаг: совершил ли пользователь хотя бы одну покупку
buyers = df_orders_AB[['user_id']].drop_duplicates()
first_purchase_df['first_purchase_flag'] = first_purchase_df['user_id'].isin(buyers['user_id']).astype(int)

# Размеры групп
n_control = first_purchase_df[first_purchase_df['group'] == 'control'].shape[0]
n_treatment = first_purchase_df[first_purchase_df['group'] == 'treatment'].shape[0]

# Количество успехов
m_control = first_purchase_df[(first_purchase_df['group'] == 'control') &
    (first_purchase_df['first_purchase_flag'] == 1)].shape[0]

m_treatment = first_purchase_df[(first_purchase_df['group'] == 'treatment') &
    (first_purchase_df['first_purchase_flag'] == 1)].shape[0]

# Z-test пропорций
stat, p_value = proportions_ztest(
    [m_control, m_treatment],
    [n_control, n_treatment],
    alternative='two-sided'
)

alpha = 0.05

print("Статистическая значимость различий:")
print(f"z-statistic = {stat:.3f}, p-value = {p_value:.4f}")

if p_value < alpha:
    print("Различие между группами статистически значимо!")
else:
    print("Различие между группами не является статистически значимым.")
````

Результаты z-теста показали p-value < 0.001, что свидетельствует о статистически значимом различии конверсии между контрольной и тестовой группами.

Таким образом, с высокой вероятностью наблюдаемый рост конверсии обусловлен именно действием бонуса, а не случайными факторами.

- Тем не менее для комплексной оценки эффекта необходимо проанализировать денежные продуктовые метрики, чтобы понять, приводит ли рост конверсии к реальному увеличению выручки.

````python
# Вспомогательные метрики
# ARPU (средняя выручка на пользователя)
revenue_per_user = df_orders_AB.groupby('user_id')['total_price'].sum().reset_index()
revenue_per_user = revenue_per_user.merge(df_users_AB[['user_id', 'group']], on='user_id')

arpu = revenue_per_user.groupby('group')['total_price'].sum() / users_per_group
arpu_change = ((arpu['treatment'] - arpu['control']) / arpu['control'] * 100).round(2)

# ARPPU (средняя выручка на покупателя)
arppu = revenue_per_user.groupby('group')['total_price'].mean()
arppu_change = ((arppu['treatment'] - arppu['control']) / arppu['control'] * 100).round(2)

# AOV (средний чек)
aov = df_orders_AB.groupby('group')['total_price'].mean()
aov_change = ((aov['treatment'] - aov['control']) / aov['control'] * 100).round(2)

# Общая выручка
total_revenue = df_orders_AB.groupby('group')['total_price'].sum()
total_revenue_change = ((total_revenue['treatment'] - total_revenue['control']) / total_revenue['control'] * 100).round(2)

print("\nВспомогательные метрики:")
print(f"\nARPU: \n{arpu.round(2)} | Изменение: {arpu_change}%")
print(f"\nARPPU: \n{arppu.round(2)} | Изменение: {arppu_change}%")
print(f"\nAOV: \n{aov.round(2)} | Изменение: {aov_change}%")
print(f"\nОбщая выручка: \n{total_revenue.round(2)} | Изменение: {total_revenue_change}%")
````

Анализ денежных метрик показал:

- ARPU вырос более чем в 10 раз (+940%), что напрямую связано с резким увеличением числа покупателей в тестовой группе.

- ARPPU увеличился на 47,5 %, что указывает на рост выручки на одного платящего пользователя.

- AOV изменился незначительно (+6.3%), следовательно бонус практически не влияет на средний чек, а действует преимущественно через рост конверсии.

- Общая выручка в тестовой группе значительно выше (+914%), что логично следует из увеличения числа покупок.

- Несмотря на выраженные различия, необходимо проверить статистическую значимость распределений выручки, поскольку денежные метрики обычно имеют сильную асимметрию и выбросы.


Для сравнения распределений выручки между группами применим непараметрический критерий Манна-Уитни, устойчивый к ненормальности распределения.

ARPU считается на уровне пользователя и имеет ненормальное распределение с большим числом нулей, поэтому применяем тест Манна-Уитни.

**Нулевая гипотеза (H0):**
ARPU в контрольной и тестовой группах не различается.

**Альтернативная гипотеза (H1):**
ARPU в контрольной и тестовой группах различается.

````python
# ARPU на уровне пользователя
revenue_all_users = df_users_AB[['user_id', 'group']].merge(
    revenue_per_user[['user_id', 'total_price']],
    on='user_id', how='left').fillna(0)

control_arpu = revenue_all_users[revenue_all_users['group'] == 'control']['total_price']
test_arpu = revenue_all_users[revenue_all_users['group'] == 'treatment']['total_price']

alpha = 0.05

p_value_arpu = mannwhitneyu(control_arpu, test_arpu, alternative='two-sided').pvalue
print(f"\nMann-Whitney test для выручки: p-value = {p_value_arpu:.4f}")

if p_value_arpu < alpha:
    print("Различие между группами статистически значимо!")
else:
    print("Различие между группами не является статистически значимым.")
````

- Критерий Манна-Уитни показал p-value < 0.05, что свидетельствует о статистически значимом различии ARPU между контрольной и тестовой группами.


- Следовательно, нулевая гипотеза об отсутствии различий отвергается. Это означает, что пользователи из тестовой группы в среднем приносят большую выручку на пользователя.


- Однако важно учитывать, что ARPU напрямую зависит от конверсии в покупку, которая ранее также показала значимый рост. Поэтому обнаруженный эффект ARPU может быть обусловлен увеличением доли платящих пользователей, а не изменением поведения самих покупателей. Для уточнения этого предположения далее анализируется метрика ARPPU.

---

ARPPU считается только по платящим пользователям и имеет асимметричное распределение, поэтому так же применим тест Манна-Уитни.

**Нулевая гипотеза (H0):**
ARPPU в контрольной и тестовой группах не различается.

**Альтернативная гипотеза (H1):**

````python
# ARPPU 
control_arppu = revenue_per_user[revenue_per_user['group'] == 'control']['total_price']
test_arppu = revenue_per_user[revenue_per_user['group'] == 'treatment']['total_price']

alpha = 0.05

p_value_arppu = mannwhitneyu(control_arppu, test_arppu, alternative='two-sided').pvalue

print(f"\nARPPU - Mann-Whitney test: {p_value_arppu:.4f}")

if p_value_arppu < alpha:
    print("Различие между группами статистически значимо!")
else:
    print("Различие между группами не является статистически значимым.")
````

- Следовательно, статистически значимых различий ARPPU между группами не обнаружено, и нулевая гипотеза не отвергается.


- Это означает, что выручка на одного платящего пользователя не изменилась, а рост ARPU объясняется прежде всего увеличением числа покупателей, а не ростом трат каждого из них.

Для проверки влияния бонуса на структуру заказов дополнительно рассмотрим метрику среднего чека (AOV).
AOV считается на уровне заказов, распределение также ненормальное - используем тест Манна-Уитни.

**Нулевая гипотеза (H0):**
AOV в контрольной и тестовой группах не различается.

**Альтернативная гипотеза (H1):**

---

Рассмотрим барьерную метрику - общую выручку от пользователей TikTok. Данная метрика является агрегированной на уровне группы, поэтому применение классических параметрических статистических тестов напрямую некорректно.


Чтобы корректно оценить различия между группами, используем бутстрап-подход. 


**Нулевая гипотеза (H0):**
Общая выручка в контрольной и тестовой группах не различается.

**Альтернативная гипотеза (H1):**
Общая выручка в контрольной и тестовой группах различается.

````python
# Фиксируем сид для воспроизводимости
np.random.seed(341)

# Выборки по группам
control = revenue_all_users[revenue_all_users['group'] == 'control']['total_price'].values
test = revenue_all_users[revenue_all_users['group'] == 'treatment']['total_price'].values

n_iterations = 1000
percentiles = [25, 50, 75]

ci_results = {}
boot_diffs_dict = {}

# Бутстрап-разница перцентилей между тестовой и контрольной группами
for p in percentiles:
    boot_diffs = []

    for i in range(n_iterations):
        boot_control = np.random.choice(control, size=len(control), replace=True)
        boot_test = np.random.choice(test, size=len(test), replace=True)

        control_p = np.percentile(boot_control, p)
        test_p = np.percentile(boot_test, p)

        # Находим разницу между значениями тестовой и контрольной групп
        boot_diffs.append(test_p - control_p)

    # Создаём массив разниц между перцентилями тестовой и контрольной групп
    boot_diffs = np.array(boot_diffs)

    # Доверительный интервал разницы между перцентилями (перцентильный метод)
    diff_ci = np.percentile(boot_diffs, [2.5, 97.5])
    boot_diffs_dict[p] = boot_diffs
    mean_diff = np.mean(boot_diffs)
    
    ci_results[p] = {
        'control': np.percentile(control, p),
        'test': np.percentile(test, p),
        'diff_ci': diff_ci
    }

# Вывод доверительного интервала
for p in percentiles:
    print(f"{p}-й перцентиль:")
    print(f"Контрольная группа: {ci_results[p]['control']:.2f}")
    print(f"Тестовая группа: {ci_results[p]['test']:.2f}")
    print(f"95%-й доверительный интервал разности [{ci_results[p]['diff_ci'][0]:.2f}, {ci_results[p]['diff_ci'][1]:.2f}]")
    print()
````

- Общая выручка в тестовой группе значительно выше (+914%), что логично следует из роста конверсии и числа покупателей.


- Поскольку метрика агрегированная на уровне группы, классические статистические тесты не применимы напрямую.


- Перцентильный анализ выручки на пользователя показал нулевые значения во всех группах, что объясняется крайне высокой долей пользователей без покупок. Следовательно, такой подход не позволяет корректно оценить барьерную метрику, поскольку общая выручка является агрегированной метрикой на уровне группы, а не распределением пользовательских значений.


- Барьерная метрика подтверждает положительный эффект бонуса через рост числа покупателей, но статистически проверить различия на уровне суммарной выручки сложно.


- В совокупности результаты барьерной метрики согласуются с положительным влиянием бонуса через рост числа покупателей, однако из-за высокой дисперсии пользовательской выручки и ограниченного объёма выборки статистическая значимость эффекта по выручке требует дополнительной проверки.
